```python
image = mpimg.imread('../resources/text.jpg')[:,:,0]
```

```python
class Segmentator:
    def __init__(self):
        self._vertical_segments = []
        self.segments = []
        self._contrast = 50
    def _vertical_segmentation(self, img):
        white_line = True
        for i in range(img.shape[0]):
            if white_line <> (np.mean(img[i]) == 255):
                white_line = not(white_line)
                self._vertical_segments.append(i)
    
                
```

```python
seg = Segmentator()
seg._vertical_segmentation(image)
```

```python
blacklined_image = np.zeros(shape=(image.shape[0],image.shape[1], 3))
for i in range(3):
    blacklined_image[:,:,i] = image
```

```python
for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        for k in range(3):
            blacklined_image[i,j,k] = 255 - blacklined_image[i,j,k]
```
